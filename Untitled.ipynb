```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from skopt import gp_minimize
from skopt.space import Real

# Drop rows where '1st melt-down time' or 'PowerCons (Kwh)' is NaN
df_cleaned_power = df.loc[df['Series'] == 400].dropna(
    subset=['VCB On Time (min)', 'PowerCons (Kwh)'])

# Plot 1st melt-down time against Power Consumption
plt.figure(figsize=(10, 6))
sns.scatterplot(x='VCB On Time (min)',
                y='PowerCons (Kwh)', data=df_cleaned_power)
plt.title('VCB On Time (min) vs Power Consumption')
plt.xlabel('VCB On Time (min)')
plt.ylabel('PowerCons (Kwh)')
plt.show()

# Define the function to minimize for 1st melt-down time


def objective_1st(meltdown_time):
    # Using linear interpolation to estimate power consumption at given meltdown time
    meltdown_time = meltdown_time[0]  # Extract scalar value from array
    power_consumption = np.interp(
        meltdown_time, df_cleaned_power['VCB On Time (min)'], df_cleaned_power['PowerCons (Kwh)'])
    return float(power_consumption)  # Ensure the return value is a scalar


# Define the search space for 1st melt-down time
search_space_1st = [Real(df_cleaned_power['VCB On Time (min)'].min(),
                         df_cleaned_power['VCB On Time (min)'].max(), name='VCB On Time (min)')]

# Run Bayesian Optimization for 1st melt-down time
result_1st = gp_minimize(objective_1st, search_space_1st,
                         n_calls=50, random_state=0)

# Extract the optimal value for 1st melt-down time
optimal_meltdown_time_1st = result_1st.x[0]
print("Optimal VCB On Time (min) (Minutes):", optimal_meltdown_time_1st)

# Visualize the result for 1st melt-down time
plt.figure(figsize=(10, 6))
sns.scatterplot(x='VCB On Time (min)', y='PowerCons (Kwh)',
                data=df_cleaned_power, label='Data Points')
plt.axvline(optimal_meltdown_time_1st, color='red', linestyle='--',
            label=f'Optimal 1st Melt-down Time: {optimal_meltdown_time_1st:.2f} min')
plt.title('1st Melt-down Time vs Power Consumption with Optimal Melt-down Time')
plt.xlabel('1st melt-down time (Minutes)')
plt.ylabel('PowerCons (Kwh)')
plt.legend()
plt.show()
```
