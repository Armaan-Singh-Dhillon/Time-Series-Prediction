# 200 series

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the slope
slope = 369.845231

# Generate x values
x = np.linspace(0, 35, 5)
# Calculate y valu3s
y = slope * x

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y)

# Add vertical lines
x1 = 13.73
x2 = 15.429
x3 = 25.98
x4 = 30.512018
plt.axvline(x=x1, color='r', linestyle='--', label='1st Optimal meltdown time')
plt.axvline(x=x2, color='b', linestyle='--', label='Average of 1st meltdown time of 200 series')
plt.axvline(x=x3, color='g', linestyle='--', label='2nd Optimal meltdown time')
plt.axvline(x=x4, color='b', linestyle='--', label='Average of 2nd meltdown time of 200 series')

# Calculate y values at x1 and x2
y1 = slope * x1
y2 = slope * x2
y3 = slope * x3
y4 = slope * x4

# Calculate the difference between the y-values
y_diff = abs(y2 - y1)
y_diff2 = abs(y3 -y4)

# Add a label displaying the y difference
midpoint = (x1+x2)/2
midpoint2 = (x3+x4)/2
plt.text(midpoint, 8000, f'Power Reduced By {
         y_diff:.2f} KWh', fontsize=12, ha='center', va='bottom', backgroundcolor='white')
plt.text(midpoint2, 12000, f'Power Reduced By {
         y_diff2:.2f} KWh', fontsize=12, ha='center', va='bottom', backgroundcolor='white')

# Add labels and title
plt.xlabel('Time(minutes)')
plt.ylabel('Power requirements')
plt.title('AI based EAF control')
plt.legend()
print(f'Toal power reduced:{y_diff+y_diff2:.2f} KWh')
# Display the plot
plt.grid(True)
plt.show()
```
